```python
plt.plot([1,4,7],[7,1,3])
plt.title("Heading")
plt.xlabel("X values")
plt.ylabel("Y values")
plt.show()
```

```python
x1 = [5,9,10]
y1 = [24,53,13]
x2 = [13,90,45]
y2 = [67,23,45]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
from matplotlib import style
```

```python
style.use('ggplot')
x1 = [5,9,10]
y1 = [24,53,13]
x2 = [13,90,45]
y2 = [67,23,45]
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.show()
```

```python
style.use('ggplot')
x1 = [5,9,10]
y1 = [24,53,13]
x2 = [13,90,45]
y2 = [67,23,45]
plt.plot(x1,y1, label="x1-y1", linewidth=3)
plt.plot(x2,y2, label="x2-y2")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x1 = [5,9,10]
y1 = [24,53,13]
x2 = [13,90,45]
y2 = [67,23,45]
plt.scatter(x1,y1, label="x1-y1", linewidth=3)
plt.plot(x2,y2, label="x2-y2")
plt.legend()
plt.show()
```

```python
style.use('ggplot')
x1 = [5,9,10]
y1 = [24,53,73]
x2 = [13,90,45]
y2 = [12,23,45]
plt.scatter(x1,y1, label="x1-y1", linewidth=3)
plt.hist(x2,y2, label="x2-y2")
plt.legend()
plt.show()
```
